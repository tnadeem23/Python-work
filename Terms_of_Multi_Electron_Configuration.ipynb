```python
def config(conf,k):
    for i in range(k,len(conf)):
        if (conf[i]=='d'):
            return '2D'
        elif (conf[i]=='s'):
            return '2S'
        elif (conf[i]=='p4'):
            return '1S1D'
        elif(conf[i]=='s2'):
            return '1S'
        elif(conf[i]=='d2'):
            return '1S1D1G'
lval = {
    'S' : 0 , 'P' : 1 , 'D' : 2 ,'F' : 3 , 'G' : 4 , 'H' : 5 , 'I' : 6 , 'K' : 7 , 'L' : 8
}
lvalopp = {
    0 : 'S' , 1 : 'P' , 2 : 'D' , 3 : 'F' , 4 : 'G' , 5 : 'H' , 6 : 'I' , 7 : 'K' , 8 : 'L'
}
        
conf =[]
print("Enter the first confguration: ")
a = input()
conf.append(str(a))
print("Enter second confguration: ")
b = input()
conf.append(str(b))
while True:
    print("Press y for adding another configuration: ")
    ch=input()
    if ch=='y':
        print("Enter second confguration: ")
        c = input()
        conf.append(str(c))
    else :
        break
obj = [[] for i in range(len(conf))]

for k in range(0,len(conf)):
    VAL= config(conf,k)
    obj[k].append(VAL)
    print("Term of " +conf[k]+ " : ", VAL)
    
term = []
for z in range(0,len(conf)):
    tm = obj[z]
    values=[]
    for i in range(0, len(tm[0]), 2):
        split=tm[0][i:i+2]
        values.append(split)
    term.append(values)
#print(term)
final = 1
for i in range(0,len(term)):
    final = final*len(term[i])
#print(final)
```

```python
#print(term[1][0])
from fractions import Fraction
for i in range(0,final):
    for j in range(i+1):
        print('Combining '+term[i][j]+' with '+term[i+1][j]+' : ')
        termchar1 = term[i][j]
        termchar2 = term[i+1][j]
        l1=lval.get(termchar1[1:])
        l2=lval.get(termchar2[1:])
        r1=int(termchar1[:1])
        r2=int(termchar2[:1])
        s1r=(r1-1)/2
        s2r=(r2-1)/2
        s1=str(Fraction(s1r))
        s2=str(Fraction(s2r))
        while True:
            S=abs(s1r+s2r)
            break
        #print(S)
        R=Fraction((2*(S)+1))
        while True:
            valL=abs(int(l1)+int(l2))
            break
        L=lvalopp.get(valL)
        J= []
        for q in range(2):
            if(q>0):
                valueJ=abs(valL-S)
            else:
                valueJ=abs(valL+S)
            valueJ=str(Fraction(valueJ))
            J.append(valueJ)
        #print(J)
        print('L1 = '+str(l1)+'\t\t\t L2 = '+str(l2))
        print('S1 = '+str(s1)+'\t\t\t S2 = '+str(s2))
        print('L  = '+str(L)+'\t\t\t R = '+str(R))
        print('\nTerm = '+str(R)+' '+str(L)+' '+J[0]+','+J[1])
        if(i==final-2):
            print('\n\nCombining '+term[i][j+1]+' with '+term[i+1][j]+' : ')
            termchar1 = term[i][j+1]
            termchar2 = term[i+1][j]
            l1=lval.get(termchar1[1:])
            l2=lval.get(termchar2[1:])
            r1=int(termchar1[:1])
            r2=int(termchar2[:1])
            s1r=(r1-1)/2
            s2r=(r2-1)/2
            s1=str(Fraction(s1r))
            s2=str(Fraction(s2r))
            while True:
                S=abs(s1r+s2r)
                break
            #print(S)
            R=Fraction((2*(S)+1))
            L=[]
            newL=[]
            for w in range(4,-1,-1):
                valL=w
                L.append(valL)
                newL.append(lvalopp.get(valL))
                valL=valL-1
            #print(newL)
            #print(L)
            J0=[]
            J1=[]
            J2=[]
            J3=[]
            J4=[]
            for q in range(5):
                if(q==0):
                    J4.append(str(Fraction(abs(L[4]+S))))
                elif(q==1):
                    J3.append(str(Fraction(abs(L[3]+S))))
                    J3.append(str(Fraction(abs(L[3]-S))))
                elif(q==2):
                    J2.append(str(Fraction(abs(L[2]+S))))
                    J2.append(str(Fraction(abs(L[2]-S))))
                elif(q==3):
                    J1.append(str(Fraction(abs(L[1]+S))))
                    J1.append(str(Fraction(abs(L[1]-S))))
                elif(q==4):
                    J0.append(str(Fraction(abs(L[0]+S))))
                    J0.append(str(Fraction(abs(L[0]-S))))
                
            print('L1 = '+str(l1)+'\t\t\t L2 = '+str(l2))
            print('S1 = '+str(s1)+'\t\t\t S2 = '+str(s2))
            print('L  = ',end="")
            for r in range(5):
                print(str(L[r])+', ',end="")
            print('\t R = '+str(R))
            print('\nTerm = ',end="")
            for r in range(5):
                print(str(R)+' '+str(newL[r])+' ',end="")
                for t in range(2):
                    if(r==0):
                        print(J0[t]+',',end="")
                    elif(r==1):
                        print(J1[t]+',',end="")
                    elif(r==2):
                        print(J2[t]+',',end="")
                    elif(r==3):
                        print(J3[t]+',',end="")
                    elif(r==4):
                        print(J4[t]+',',end="")
                        break
                    if(t>=1):
                        print('\n')
                
        
    break
```
