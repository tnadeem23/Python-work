```python
import numpy as np
import pandas as pd
from random import randint

inventory,k=100,32
demand,z,arrival,n,timebwdemand,demandsize,cost=0,0,0,0,0,0,0
print('Product Inventory System')
n=int(input('Enter number of months : '))
print('Initial Inventory Level = ' ,inventory)
print('Total Number of Demands = ',n)
print('Length of Simulation = ' ,n)
print("k=32 , i=3")

df = pd.DataFrame(columns=['Order ID','Inventory Level','Time Between Demands','Demand Size','Status','Items to Order','Cost','Arrival Time'])
print(df)

for i in range(n):
    timebwdemand = randint(15,21)
    demand = randint(0,11)
    if(demand==1 or demand==2):
        demandsize=10
    elif(demand==3 or demand==4 or demand==5):
        demandsize=20
    elif(demand==6 or demand==7 or demand==8 or demand==9 or demand==10):
        demandsize=40
        
    if(demandsize<=inventory):
        status='Success'
    elif(demandsize>inventory):
        status='Fail'
        z=demandsize-inventory
        if(z==0):
            inventory=0
        cost=k+(i*z)
        arrival=randint(15,31)
    df = df.append(pd.Series([i+1, inventory, timebwdemand, demandsize, status, z, cost, arrival], index=df.columns ), ignore_index=True)
    inventory = inventory - demandsize
    if (inventory<0):
        inventory = 0;
print(df)
```
